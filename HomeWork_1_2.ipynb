# Задание №1: Работа в Pandas, на примере задачи "Titanic"

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AlexeyTri/Home_work_NN/refs/heads/main/data/titanic.csv", index_col='PassengerId')
df.head()
```

```python
df.info()
```

1. Какое число мужчин и женшин ехало на корабле: 577мужчин, 314 женщин

```python
df['Sex'].value_counts()
```

2. Какой части пассажиров удалось выжить? Рассчитать долю выживших пассажиров: 38%

```python
servived = np.round(df['Survived'].sum()/df.shape[0], 2) * 100
servived.item()
```

3. Какую долю пассажиры первого класса составляли среди всех пассажиров?

```python
class1sur = np.round(df.groupby(['Pclass'])['Survived'].count()[1] / df.shape[0], 2)*100
class1sur.item()
```

4. Какого возраста были пассажиры? Посчитайте среднее и медиану
возраста пассажиров:

```python
print(f"Средниц возраст пассажиров: {np.round(df['Age'].mean(), 2)}, медианный возраст пассажиров: {df['Age'].median()}")
```

5. Коррелируют ли число братьев/сестер с числом родителей/детей?
Посчитайте корреляцию Пирсона между признаками SibSp и Parch: корреляция низка, составляет 0.41

```python
df[['SibSp', 'Parch']].corr(method='pearson')
```

6. Какое самое популярное женское имя на корабле?: "Andersson"

```python
fem = df[df['Sex'] == 'female'].reset_index(drop=True)['Name']
```

```python
fem.str.split(r",", expand=True)[0].value_counts()
```

```python
df['Sex'] = df['Sex'].replace(['male', 'female'], [0, 1])
```

```python
y = df['Survived'].values
y
```
