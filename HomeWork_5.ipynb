```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.preprocessing import scale, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.linear_model import Perceptron
```

```python
X_train_load = pd.read_csv("https://raw.githubusercontent.com/AlexeyTri/Home_work_NN/refs/heads/main/data/perceptron-train.csv")
X_test_load = pd.read_csv("https://raw.githubusercontent.com/AlexeyTri/Home_work_NN/refs/heads/main/data/perceptron-test.csv")
```

```python
y_train = X_train_load.iloc[:, 0]
y_test = X_test_load.iloc[:, 0]
X_train = X_train_load.iloc[:, 1:].values
X_test = X_test_load.iloc[:, 1:].values
```

```python
perceptron = Perceptron(random_state=241)
perceptron.fit(X_train, y_train)
accuracy_score(y_test, perceptron.predict(X_test)), accuracy_score(y_train, perceptron.predict(X_train))
```

```python
std_scl = StandardScaler()
X_train_scale = std_scl.fit_transform(X_train)
X_test_scale = std_scl.transform(X_test)
```

```python
perceptron.fit(X_train_scale, y_train)
accuracy_score(y_test, perceptron.predict(X_test_scale)), accuracy_score(y_train, perceptron.predict(X_train_scale))
```

```python
accuracy_score(y_test, perceptron.predict(X_test_scale)) - accuracy_score(y_test, perceptron.predict(X_test))
```
