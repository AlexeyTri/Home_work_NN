```python
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
```

```python
df_price = pd.read_csv("https://raw.githubusercontent.com/AlexeyTri/Home_work_NN/refs/heads/main/data/close_prices.csv")
df_index = pd.read_csv("https://raw.githubusercontent.com/AlexeyTri/Home_work_NN/refs/heads/main/data/djia_index.csv")
```

```python
df_price.drop(['date'], axis=1, inplace=True)
```

```python
pca = PCA(n_components=10)
pca_matrix = pca.fit_transform(df_price)
```

```python
pca.explained_variance_ratio_, pca.explained_variance_ratio_[0]+pca.explained_variance_ratio_[1]+pca.explained_variance_ratio_[2]+pca.explained_variance_ratio_[3]
```

```python
max_index = pd.DataFrame(abs(pca.components_[0]), columns=['value']).sort_values(by=['value'], ascending=False).index[0]
df_price.iloc[:, max_index]
```

```python
np.corrcoef(pca_matrix[:, 0], df_index)
```

```python
np.corrcoef(pca_matrix[:, 0], df_index.iloc[:, 1].values)
```
