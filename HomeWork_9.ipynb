```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.metrics import roc_auc_score, accuracy_score, precision_score, recall_score, f1_score, precision_recall_curve
from sklearn.linear_model import LogisticRegression
```

```python
df_classification = pd.read_csv("https://raw.githubusercontent.com/AlexeyTri/Home_work_NN/refs/heads/main/data/classification.csv")
df_scores = pd.read_csv("https://raw.githubusercontent.com/AlexeyTri/Home_work_NN/refs/heads/main/data/scores.csv")
```

```python
!pip install pycm
import pycm
from pycm import ConfusionMatrix
y_true = df_classification['true'].values
y_prediction = df_classification['pred'].values
cm = ConfusionMatrix(actual_vector=y_true,predict_vector=y_prediction)
print(cm)
```

```python
cm.matrix.get(1).get(1), cm.matrix.get(1).get(0), cm.matrix.get(0).get(0), cm.matrix.get(0).get(1)
```

```python
print(f"Accuracy_score: {accuracy_score(df_classification['true'], df_classification['pred'])} {'\n'} Precision_score: {precision_score(df_classification['true'], df_classification['pred']):.2f} {'\n'} recall_score: {recall_score(df_classification['true'], df_classification['pred']):.2f} {'\n'} f1_score: {f1_score(df_classification['true'], df_classification['pred']):.2f} {'\n'}")
```

```python
roc_auc_score(df_scores['true'], df_scores['score_logreg']), roc_auc_score(df_scores['true'], df_scores['score_svm']), roc_auc_score(df_scores['true'], df_scores['score_knn']), roc_auc_score(df_scores['true'], df_scores['score_tree'])
```

```python
pres_logreg = precision_recall_curve(df_scores['true'], df_scores['score_logreg'])[0]
recall_logreg = precision_recall_curve(df_scores['true'], df_scores['score_logreg'])[1]
prc_logreg = pd.DataFrame(np.vstack([pres_logreg, recall_logreg]).T, columns=(['prec', 'recall']))
prc_logreg[prc_logreg['recall'] > 0.7]['prec'].max()
```

```python
pres_svm = precision_recall_curve(df_scores['true'], df_scores['score_svm'])[0]
recall_svm = precision_recall_curve(df_scores['true'], df_scores['score_svm'])[1]
prc_svm = pd.DataFrame(np.vstack([pres_svm, recall_svm]).T, columns=(['prec', 'recall']))
prc_svm[prc_svm['recall'] > 0.7]['prec'].max()
```

```python
pres_knn = precision_recall_curve(df_scores['true'], df_scores['score_knn'])[0]
recall_knn = precision_recall_curve(df_scores['true'], df_scores['score_knn'])[1]
prc_knn = pd.DataFrame(np.vstack([pres_knn, recall_knn]).T, columns=(['prec', 'recall']))
prc_knn[prc_knn['recall'] > 0.7]['prec'].max()
```

```python
pres_tree = precision_recall_curve(df_scores['true'], df_scores['score_tree'])[0]
recall_tree = precision_recall_curve(df_scores['true'], df_scores['score_tree'])[1]
prc_tree = pd.DataFrame(np.vstack([pres_tree, recall_tree]).T, columns=(['prec', 'recall']))
prc_tree[prc_tree['recall'] > 0.7]['prec'].max()
```
